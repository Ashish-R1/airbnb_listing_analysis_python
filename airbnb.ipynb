# Handling missing values

```python
df['last review'] = pd.to_datetime(df['last review'], errors = 'coerce')
```

```python
df.info()
```

```python
df.head()
```

```python
df.fillna({'reviews per month':0, 'last review': df['last review'].min()}, inplace = True)
```

```python
df.dropna(subset = ['NAME','host name'], inplace = True)
```

```python
print(df.isnull().sum())
```

```python
df = df.drop(columns = ['license','house_rules'], errors = 'ignore')
```

```python
df.head()
```

# Removing symbols

```python
df['price'] = df['price'].replace('[\$,]','', regex = True).astype(float)
df['service fee'] =df['service fee'].replace('[\$,]','', regex = True).astype(float)
```

```python
df.head()
```

# Remove duplicates

```python
df.drop_duplicates(inplace = True)
```

```python
df.info()
```

# Descriptive Statistics

```python
df.describe()
```

```python
df.groupby(by='neighbourhood group')['review rate number'].mean().sort_values(ascending=False)
```

# Distribution of listing prices

```python
plt.figure(figsize = (10,5))
sns.histplot(df['price'], bins = 50, kde= True, color ='red')
plt.title('Distribution of listing price')
plt.xlabel('Price $')
plt.ylabel('Frequency')
plt.show()
```

# Distribution of different room types

```python
df['room type']
```

```python
plt.figure(figsize = (8,5))
sns.countplot(x='room type', data = df, color  = 'blue')
plt.title("Room type distribution")
plt.xlabel("room type")
plt.ylabel("count")
plt.show()
```

# Listing distribution across different neighbourhood

```python
plt.figure(figsize = (12,5))
sns.countplot(y='neighbourhood group', data = df, color  = 'lightgreen', order = df['neighbourhood group'].value_counts().index)
plt.title("Number of listing in neighbourhood group")
plt.xlabel("Count")
plt.ylabel("neighbourhood group")
plt.show()
```

# Price dependency on  Neighbourhood group

```python
mean_neighbourhood_price = df.groupby(by='neighbourhood group')['price'].mean().sort_values(ascending=False)
mean_neighbourhood_price
```

```python
plt.figure(figsize=(10, 5))
sns.barplot( x=mean_neighbourhood_price.index, y=mean_neighbourhood_price.values,color='skyblue')
plt.title('Average Airbnb Price by Neighbourhood Group', fontsize=16)
plt.xlabel('Neighbourhood Group', fontsize=12)
plt.ylabel('Average Price (USD)', fontsize=12)
```

# Relation between price and room type

```python
plt.figure(figsize = (10,6))
sns.boxplot(x='room type',y='price', data = df, hue  = 'room type', palette = 'Set1')
plt.title("price vs room type")
plt.xlabel("room type")
plt.ylabel("Price $")
plt.legend(title = 'Room type')
plt.show()
```

# Number of reviews changed over time

```python
df['last review']= pd.to_datetime(df['last review'])
reviews_over_time = df.groupby(df['last review'].dt.to_period('M')).size()

plt.figure(figsize=(12,6))
reviews_over_time.plot(kind= 'line', color='red')
plt.xlim(right='2025-01')
plt.title('Number of reviews over time')
plt.xlabel('Date')
plt.ylabel('Number of Reviews')
plt.show()
```
